# Clasificaci√≥n Binaria

## Conceptos importantes   

**Separabilidad Lineal**
* La separabilidad lineal es una propiedad del conjunto de datos con el que estamos trabajando y no del modelo que vamos a entrenar.
* Un conjunto es linealmente separable si puedo encontrar al menos una recta que me separe los dos grupos.

**Clasificadores lineales**    

* Las entradas son vectores de n componenetes (x,y)
* Los clasifificadores pueden ser divididos en 2 etapas:    
 **1) Funci√≥n de predicci√≥n** -> arroja un valor num√©rico sobre el cual luego tomo una decisi√≥n dependiendo del umbral de decisi√≥n ue haya fijado.   
 **f(x) = W(t) X +W0**    

 **W(t)=** Producto punto de un vector de peso   
 **X=** Vector de datos    
 **W(t) X=** Representa la suma de W1*X1 + W2*X2 + W3*X3......+Wn*Xn
 **W0=** Par√°metro de sesgo (Bias)   
 
 **2) Funci√≥n de decisi√≥n** -> Seg√∫n el signo del valor num√©rico obtenido en la funci√ßon de predicci√≥n, tomo una decisi√≥n. La decisi√≥n la fijo arbitrariamente a priori.

## Algoritmos de clasificaci√≥n Binaria
### Clasificaci√≥n con el Algoritmo del Perceptr√≥n
* Es un algoritmo desarrollado para clasificar datos en dos categor√≠as (1 √≥ -1), si es que estos son linealmente separables.
* El objetivo es encontrar un hiperplano de separaci√≥n. Un hiper plano es una generalizaci√≥n de una recta a un espacio de n dimensiones. (1 dimension -> recta; 2 dimensiones -> plano; 3 o m√°s dimensiones -> hiperplanos)  
* Este clasificador lineal binario separa el espacio de los datos en 2, esta separacion la hace mediante una frontera. A un lado de la forntera quedan los positivos (1) y al otro lado los negativos (-1).

#### Entrada:
* **Una secuencia de pares de entrenamientos** (x1, y1; x2, Y2....)
* **Un hiperpar√°metro** = Tasa constante de aprendizaje **r** (n√∫mero muy peque√±o y menor a 1)
* Para cada par de datos genero una predicci√≥n seg√∫n mi modelo (+1, -1) y al resultado de esa predicci√≥n lo comparo con lo que tengo en el conjunto de entrenamiento:
* * Si mi prediccion es correcta, no hago nada. 
* * Si mi prediccion es incorrecta actualizo el vector de pesos de tal forma que sea igual a la versi√≥n de la iteraci√≥n anterior + la constante r * el producto del vector x,y (W(t+1)= W(t)+r*(x,y))     
* Nuestras incognitas, es decir, lo que buscamos aprender es el valor de los coeficientes de W y W0. Para encontrar una funci√≥n que me separe dos nubes de puntos.
* Si w0=0 -> La recta para separa el hiperplano pasa por el origen. En el algoritmo W0 es al azar o le puedo fijar la semilla a W0.

### Evoluci√≥n del hiperplano cuando se comete un error

* En el gr√°fico anterior, podemos ver que el producto punto del vector (x,y) * W va a ser negativo, ya que W y el vector(x,y) apuntan para lados opuestos. Pero la predicci√≥n dice que debe ser clasificado como positivo -> estoy cometiendo un **error de predicci√≥n**.
* La ecuaci√≥n de actualizaci√≥n me dice que debo sumarle al vector de pesos W la muestra que estoy clasificando mal por un factor de actualizaci√≥n r (W(t+1)= W(t)+r*(x,y)).
* En el gr√°fico y= r   
* El vector W es normal al hiperplano... por eso cuando se mueve el vector W, el hiperplano se desplaza y se actualiza su posici√≥n
* El algoritmo hace esta correcci√≥n por cada par de datos que va analizando   

### Criterio de parada del algoritmo
* Puede ser que decida que una vez que no cometa m√°s errores y ya no se actualice m√°s el hiperplano ahi decida parar de correr el algoritmo..... esto es algo bastante hipot√©tico.
* En la realidad yo nunca s√© si un conjunto es linealmente separable. Por lo tanto en la pr√°ctica el **criterio de parada** suele ser el **n√∫mero de iteraciones** √≥ el n√∫mero de pasadas completas por el conjunto de datos, esa pasada completa se denomina **√©poca**.
* Hay que tener en cuenta que cada vez que yo actualizo el hiperplano debo volver a pasar por todo el cunto otra vez, para chequear que esa actualizaci√≥n no haya generado nuevos errores. Por eso tengo que definir un n√∫mero de √©pocas alto.

### Suponiendo que nuestros datos son linealmente separables...

* Como vemos hay infinitas soluciones al problema de clasificaci√≥n
* Todas las rectas que separen estos puntos son posibles soluciones que puede encontrar el algoritmo del perceptr√≥n.
* La mejor soluci√≥n es aquella que posea el **margen m√°ximo**: es decir que genera el hiperplano m√°s estable ante perturbaciones de la entrada. Es la soluci√≥n que tiene m√°s distancia entre una nube de puntos y la frontera de decisi√≥n.

* * Uno vs. todos
* * todos vs. todos
